```python
class Invoice:

    def calculate(self, price, gst):

        gst_amount = price * gst / 100

        total = price + gst_amount

        print("GST:", gst_amount)
        print("Invoice Amount:", total)

invoice = Invoice()

invoice.calculate(10000, 18)
```

```python
#Calculate salary and bonus and provide the total.
class Salary:

    def calculate(self, salary):

        bonus = salary * 0.10

        total = salary + bonus

        print("Bonus:", bonus)
        print("Total Salary:", total)

employee = Salary()

employee.calculate(50000)
```

```python
# display subfields using class and functions
class SubfieldsInAI:

    def Subfields(self):

        ai_fields = [
            "Machine Learning",
            "Neural Networks",
            "Vision",
            "Robotics",
            "Speech Processing",
            "Natural Language Processing"
        ]

        print("Sub-fields in AI are:")

        for field in ai_fields:
            print(field)


ai = SubfieldsInAI()
ai.Subfields()
```

```python
# Function to check marriage eligibility

def marriage_eligibility(gender, age):

    if gender == "M":
        if age >= 21:
            print("Eligible for Marriage")
        else:
            print("Not Eligible for Marriage")

    elif gender == "F":
        if age >= 18:
            print("Eligible for Marriage")
        else:
            print("Not Eligible for Marriage")

    else:
        print("Invalid Gender")


# User Input
gender = input("Enter Gender (M/F): ").upper()
age = int(input("Enter Age: "))

# Function Call
marriage_eligibility(gender, age)
```

```python
#odd or even number
def odd_even(number):

    if number % 2 == 0:
        return "Even Number"
    else:
        return "Odd Number"


num = int(input("Enter a Number: "))

result = odd_even(num)

print(result)
```
