```python
#Calculate Salary.
def calculate_salary(monthly_salary):
    bonus = monthly_salary * 0.10
    total_salary = monthly_salary + bonus
    return total_salary

salary = float(input("Enter Monthly Salary: ₹"))

final_salary = calculate_salary(salary)

print("Bonus (10%): ₹", salary * 0.10)
print("Total Salary: ₹", final_salary)
```

```python
#Invoice Calculator.
def calculate_invoice(product_price, gst_percent):
    gst = product_price * gst_percent / 100
    final_amount = product_price + gst
    return final_amount

price = float(input("Enter Product Price: ₹"))
gst = float(input("Enter GST Percentage: "))

invoice = calculate_invoice(price, gst)

print("Final Invoice Amount: ₹", invoice)
```

```python
# withdrawal ATM from balanace.
def withdraw(balance, amount):
    if amount <= balance:
        balance = balance - amount
        return balance
    else:
        return "Insufficient Balance"

balance = float(input("Enter Current Balance: ₹"))
withdraw_amount = float(input("Enter Withdrawal Amount: ₹"))

remaining = withdraw(balance, withdraw_amount)

print("Result:", remaining)
```

```python
#X Student Result
def student_result(english, tamil, maths, science,social):
    total = english + tamil + maths + science + social
    average = total / 5

    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 50:
        grade = "C"
    else:
        grade = "Fail"

    return total, average, grade

english = int(input("Enter English Mark: "))
tamil = int(input("Enter Tamil Mark: "))
maths = int(input("Enter Maths Mark: "))
science = int(input("Enter Science Mark: "))
social = int(input("Enter Social Mark: "))

total, average, grade = student_result(english, tamil, maths, science, social)

print("Total:", total)
print("Average:", average)
print("Grade:", grade)
```

```python
#Leave approval system

def leave_request(employee_name, leave_days):
    if leave_days <= 5:
        status = "Leave Approved"
    else:
        status = "Manager Approval Required"

    return employee_name, status

name = input("Enter Employee Name: ")
days = int(input("Enter Number of Leave Days: "))

employee, result = leave_request(name, days)

print("Employee:", employee)
print("Status:", result)
```
